```python
import matplotlib.pyplot as plt

# So 500 tocki random zemeni niz celiot interval
x = np.linspace(0, 2 * np.pi, 500)

A1, a1 = 1, 1
v1 = A1 * np.sin(a1 * x)

A2, a2 = 0.5, 2.5
v2 = A2 * np.sin(a2 * x)

plt.plot(x, v1)
plt.plot(x, v2)
plt.show()

import numpy as np
#se iscrtuva nula vektorot
# 0 * sin(1 * x) e nulta func vo site 500 tocki
v0 = 0 * np.sin(1 * x)

plt.figure()
plt.title("Нулта функција")
plt.plot(x, v0)
plt.grid(True)
plt.show()

# skalarni vektori definirame
k1 = 2      
k2 = -0.5   

v1_k1 = k1 * v1
v2_k2 = k2 * v2

plt.figure()
plt.title("Скалирање на вектори")
plt.plot(x, v1, label="v1")
plt.plot(x, v1_k1, label="2 * v1")
plt.plot(x, v2, label="v2")
plt.plot(x, v2_k2, label="-0.5 * v2")
plt.legend()
plt.grid(True)
plt.show()

# Zbir na vektori
v_sum = v1 + v2          

plt.figure()
plt.title("Збир v1 + v2")
plt.plot(x, v1, label="v1")
plt.plot(x, v2, label="v2")
plt.plot(x, v_sum, label="v1 + v2")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#r2 <- r2 + λ * r1
import numpy as np

n = 4  # ред на матрицата

# kvad mat a so br od 0
A = np.random.randint(low=0, high=10, size=(n, n))
print("A =\n", A, "\n")

# Копија на A
B = np.array(A)

# Бираме две различни редици и скалар λ
r1 = 0       # индекс на првата редица
r2 = 1       # индекс на втората редица
lam = 3      # скалар λ (исто цел број)

# Редична операција: r2 <- r2 + λ * r1
B[r2, :] = B[r2, :] + lam * B[r1, :]

print("B =\n", B, "\n")

# Детеминанти
detA = np.linalg.det(A)
detB = np.linalg.det(B)

print("det(A) =", detA)
print("det(B) =", detB)
```

```python
#r1 <- λ * r1
import numpy as np

n = 4  # ред на матрицата

# Квадратна матрица A со случајни цели броеви 0–9
A = np.random.randint(low=0, high=10, size=(n, n))
print("A =\n", A, "\n")

# Копија од A
B = np.array(A)

# Бираме редица r1 и скалар λ
r1 = 0       # индекс на редицата (прва редица)
lam = 3      # скалар λ

# Редична операција: r1 <- λ * r1
B[r1, :] = lam * B[r1, :]

print("B =\n", B, "\n")

# Детеминанти
detA = np.linalg.det(A)
detB = np.linalg.det(B)

print("det(A) =", detA)
print("det(B) =", detB)
print("det(B) / det(A) =", detB / detA)
```

```python
# Размена на редиците r1 и r2
import numpy as np

n = 4  # ред на матрицата

# kvad mat so slucajni broevi od 0 do 9
A = np.random.randint(low=0, high=10, size=(n, n))
print("A =\n", A, "\n")

# kopija na A
B = np.array(A)

# dvete redici
r1 = 0
r2 = 1

# zamena na redicite
tmp = np.array(B[r1, :])
B[r1, :] = B[r2, :]
B[r2, :] = tmp

print("B =\n", B, "\n")

# det
detA = np.linalg.det(A)
detB = np.linalg.det(B)

print("det(A) =", detA)
print("det(B) =", detB)
print("det(B) / det(A) =", detB / detA)
```

```python
# Транспонирана матрица B = A^T
import numpy as np

n = 4 #od koj red e

# kvad matrica A so celi broevi
A = np.random.randint(0, 10, size=(n, n))
print("A =\n", A, "\n")

#transponiranata B=A^T
B = A.T
print("B = A^T =\n", B, "\n")

#determinantite
detA = np.linalg.det(A)
detB = np.linalg.det(B)

print("det(A) =", detA)
print("det(B) =", detB)
```

```python
#vazi detab=deta *detb
import numpy as np

n = 4

A = np.random.randint(0, 10, size=(n, n))
B = np.random.randint(0, 10, size=(n, n))

detA = np.linalg.det(A)
detB = np.linalg.det(B)

AB = A @ B          
detAB = np.linalg.det(AB)

print("det(A) =", detA)
print("det(B) =", detB)
print("det(AB) =", detAB)
print("det(A)*det(B) =", detA * detB)
```

```python
##FURIEVATA
import numpy as np
import matplotlib.pyplot as plt

#Vremenski domen, 1 sek so cekor 0.001 s (fs = 1000 Hz)
t = np.arange(0, 1, 0.001)

# Cist sinusoiden signal, komv od 5 i 20 herz
clean_signal = np.sin(2 * np.pi * 5 * t) + 0.5 * np.sin(2 * np.pi * 20 * t)

# dodavame nekoj rand shum
signal = clean_signal + np.random.normal(0, 1, len(t))

# furieva
freq_domain = np.fft.fft(signal)
freq = np.fft.fftfreq(len(t), 0.001)  # 0.001 е временскиот чекор (dt)

plt.subplot(2, 1, 1)
plt.plot(t, signal)
plt.title("Original Signal in Time Domain")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")

plt.subplot(2, 1, 2)
plt.plot(freq, np.abs(freq_domain))
plt.title("Signal in Frequency Domain")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude")
plt.xlim(0, 50)  
plt.tight_layout()
plt.show()
```

```python
#Следно, користејќи ја библиотеката scipy, вчитај произволна wav гласовна датотека (аудио) и одреди го фреквентниот домен.
#from scipy.io import wavfile

import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile

#za da se najde kaj e grdiov wav fajl
#import os
#for root, dirs, files in os.walk('/'):
#    if 'meow.wav' in files:
#        print("GO NAJDOV:", os.path.join(root, 'meow.wav'))


# pat do wavot
file_path = "/drive/notebooks/meow.wav"    

# se cita audio
sampling_rate, data = wavfile.read(file_path)
print("Sampling rate:", sampling_rate, "Hz")
print("Shape of data:", data.shape)

# ako e so dva kanali nekoj edge case
if len(data.shape) > 1:
    data = data[:, 0]

# furieva
N = len(data)
fft_data = np.fft.fft(data)
frequencies = np.fft.fftfreq(N, d=1 / sampling_rate)

# crtame poz frekv
plt.figure(figsize=(12, 6))
plt.plot(frequencies[:N // 2], np.abs(fft_data)[:N // 2])
plt.title("Frequency Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid()
plt.show()
```

```python
import kagglehub
import os
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# prezemuvame .csv od kaggle
path = kagglehub.dataset_download("dineshpiyasamara/sentiment-analysis-dataset")
print("Path to dataset files:", path)
print("Files there:", os.listdir(path))

# go vcituvame datasetot od .csv
data = pd.read_csv(path + '/sentiment_analysis.csv', index_col='id')

# za da gi podelime tvitovite na pos i neg
pos = []   # sentiment = 0
neg = []   # sentiment = 1

for i, row in data.iterrows():  
    if row.iloc[0] == 0:        # 0 e pos tvit
        pos.append(row.iloc[1])
    else:                       # 1 e neg tvit
        neg.append(row.iloc[1])

print("Број на позитивни твитови:", len(pos))
print("Број на негативни твитови:", len(neg))

# zemame 10 pos 10 neg za klasterizacija
documents = pos[:10] + neg[:10]
print("Вкупно документи:", len(documents))

#tf idf vektorizacija rabotava
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(documents)

# KMeans, 2 klasteri imame pos i neg
k = 2  
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X)

labels = kmeans.labels_

#prikazuvanje
for i, doc in enumerate(documents):
    print(f"Document {i + 1}: Cluster {labels[i]} -> {doc[:100]}")
```
